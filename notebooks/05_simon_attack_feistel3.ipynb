## Derived Simon circuit

Build the Simon instance induced by the Feistel reduction and draw its circuit.

```python
from simon_qcrypto.algorithms.simon.quantum import build_circuit
from simon_qcrypto.attacks.feistel3.oracle import build_oracle

oracle = build_oracle(half_bits=3, seed=17)
circuit = build_circuit(oracle.as_simon_problem(), measure=True)
circuit.draw('mpl', fold=-1, idle_wires=False)
```
